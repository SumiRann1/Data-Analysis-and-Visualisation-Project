#USED TO GET PERCENTAGE AND GRADE SYSTEMS

```python
df["percentage"] = (df["CSL100"] + df["PHL100"] + df["CYL100"] + df["MAL100"] + df["NSO/NSS"] + df["LAN100"])/6
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
```

```python
df["percentage_std"] = std.fit_transform(df[["percentage"]])
df
```

```python
df.describe()
```

```python
count , bins, patches = plt.hist(df["percentage_std"],bins =9, edgecolor = "black")
for i in range(9):
    if i == 0:
        patches[i].set_facecolor("red")
    elif i > 5:
        patches[i].set_facecolor("green")
plt.xlabel("percent")
plt.show()
print(count)
for patch in bins:
    print(patch, end="---")
```
